Define the analytic derivative of the function`f` here (again, it should work both for single numbers and numpy arrays).

```python
def df_analytic(x):
    return -numpy.sin(x);
```

```python
assert numpy.isclose(df_analytic(0.5), -0.4794255386)
```

Now define the numerical derivative using the backward difference method. The function `backward_difference` takes three arguments, `f`, the function to calculate the derivative for, `x` the position to calculate the derivative at and `dx` the interval length.

```python
def backward_difference(f, x, dx):
    return ((f(x)-f(x-dx))/dx);
```

```python
assert numpy.isclose(
    backward_difference(f, 0.1, 0.0001),
    -0.09978366627194646
)
```

The numerical method loses accuracy due to numerical precision being limited. As the computer cannot store an infinite number of decimal places, the computer systematically recalculates the value of $\Delta$x. Therefore, there is inevitably an error in the produced value. The error in the forward/backwards difference is roughly proportional to $\Delta$x. For the central difference, the error varies significantly with $\Delta$x as it has a high polynomial dependency.
