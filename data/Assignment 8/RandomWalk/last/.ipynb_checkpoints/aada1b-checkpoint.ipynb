## Random Walks

This assignment will look at some properties of random walks.

```python
import numpy
from matplotlib import pyplot as plt 
```

To do our work we will implement a walker class. When initialised a list of possible steps is populated. In one dimension it is

[+s] , [-s] 

where s is the step size, it defaults to 1 but can be set as an argument in the constructor. In two dimensions the steps list contains

[ +s , 0 ] , [ -s , 0 ] ,  [ 0 , +s ] , [ 0 , -s ]

At each step the current position of the walker, saved in `self.pos`, is updated by adding one of the possible steps. The function `pickStep` chooses randomly one of the possible steps. Use this function to implement the `doSteps` function that performs `n` steps and returns a `(n+1) x ndim` array representing the trajectory of the walker, including the starting point. 

```python

class walker:
    def __init__(self,x0,ndim=1, step_size=1.0):
        self.pos=x0
        self.ndim=ndim
        self.possibleSteps=[]
        for i in range(ndim):
            step=numpy.zeros(ndim)
            step[i]= - step_size
            self.possibleSteps.append(numpy.array(step,dtype='f'))
            step[i]= + step_size
            self.possibleSteps.append(step.copy())
        self.npossible=len(self.possibleSteps)

    def pickStep(self):
        istep = numpy.random.choice(range(self.npossible))
        return self.possibleSteps[istep]
        
    def doSteps(self,n):
        positions=numpy.ndarray((n+1,self.ndim),dtype='f')
        # YOUR CODE HERE:
        positions[0] = 0.
        x0 = positions[0]
        for i in range(len(positions)-1):
            self.__init__(x0,self.ndim, step_size=1.0)
            positions[i+1] = positions[i] + self.pickStep()
        return positions
```

```python
# this test is worth 2 marks
numpy.random.seed(1111)
w = walker(numpy.zeros(1))
pos_test = w.doSteps(10)
reference = [[ 0.], [-1.], [ 0.], [ 1.], [ 2.], [ 1.], [ 0.], [-1.], [-2.], [-3.], [-4.]]
assert len(pos_test)==11
# plots to help debugging
plt.plot(range(11),pos_test, label='your trajectory')
plt.plot(range(11),reference,'o', label='reference points')
plt.legend()
plt.xlabel('step number')
assert (pos_test == reference).all()
```

**Task 1**

Make a plot of average position and average squared position of 100 1D walkers using 1000 steps. Your plot needs a legend, title and labels. [5 marks]


```python
nsteps = 1000
ndim = 1
nwalkers = 100
ys = w.doSteps(nsteps)
yss = numpy.square(w.doSteps(nsteps))
for i in range(nwalkers):
    w = walker(numpy.zeros(1))
    a = w.doSteps(nsteps)
    ys += a
    yss += numpy.square(a)
    
plt.plot(range(nsteps+1),ys,label = 'average')
plt.plot(range(nsteps+1),yss,label = 'average squared')

plt.legend()
plt.ylabel('position')
plt.xlabel('step')
plt.title('position of 1000 walkers')
```

Make a plot to show that the average squared distance scaling is independent of the dimension in which the walker moves. Use 100 steps and 400 walkers and use $D=1,2,3,4$. The plot should have a title, legend and labels. [5 marks]


```python
nsteps = 100
nwalkers = 400
dimensions = range(1,5)
ys = numpy.square(w.doSteps(nsteps))
for d in dimensions:
    ndim = d
    for i in range(400):
        w = walker(numpy.zeros(1))
        ys += numpy.square(w.doSteps(nsteps))
    plt.plot(range(nsteps+1),ys, label = d)
plt.legend()
plt.ylabel('distance')
plt.xlabel('steps')
plt.title('distance in each dimension')
```
