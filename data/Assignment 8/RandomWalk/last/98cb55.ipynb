To do our work we will implement a walker class. When initialised a list of possible steps is populated. In one dimension it is

[+s] , [-s] 

where s is the step size, it defaults to 1 but can be set as an argument in the constructor. In two dimensions the steps list contains

[ +s , 0 ] , [ -s , 0 ] ,  [ 0 , +s ] , [ 0 , -s ]

At each step the current position of the walker, saved in `self.pos`, is updated by adding one of the possible steps. The function `pickStep` chooses randomly one of the possible steps. Make a plot to show that the average squared distance scaling is independent of the dimension in which the walker moves. Use 100 steps and 400 walkers and use $D=1,2,3,4$. The plot should have a title, legend and labels. [5 marks]


```python
nsteps = 100
nwalkers = 400
dimensions = range(1,5)
for x in dimensions:
    y1=numpy.zeros((nsteps+1,x))
    y2=numpy.zeros((nsteps+1,x))
    d=numpy.zeros((nsteps+1,1))
    for y in range(nwalkers):
        z=walker(numpy.zeros(x),ndim=x).doSteps(nsteps)
        for t in range(1,len(z)):
            y1[t]+=z[t]
            y2[t]+=(z[t])**2
            for s in z[t]:
                p=0
                p+=s**2
            d[t]+=p
    d/=nwalkers
    plt.plot(range(0,nsteps+1),d,label='dimensions'+str(x))

plt.ylabel('Average squared distance of walkers')  
plt.xlabel('Number of steps')
plt.title('Graph showing the average squared distance for walkers with different numbers of dimensions')
plt.legend()
```
