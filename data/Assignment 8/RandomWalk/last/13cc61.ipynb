## Random Walks

This assignment will look at some properties of random walks.

```python
import numpy
from matplotlib import pyplot as plt 
```

To do our work we will implement a walker class. When initialised a list of possible steps is populated. In one dimension it is

[+s] , [-s] 

where s is the step size, it defaults to 1 but can be set as an argument in the constructor. In two dimensions the steps list contains

[ +s , 0 ] , [ -s , 0 ] ,  [ 0 , +s ] , [ 0 , -s ]

At each step the current position of the walker, saved in `self.pos`, is updated by adding one of the possible steps. The function `pickStep` chooses randomly one of the possible steps. Use this function to implement the `doSteps` function that performs `n` steps and returns a `(n+1) x ndim` array representing the trajectory of the walker, including the starting point. 

```python
import numpy as np
class walker:
    def __init__(self,x0,ndim=1, step_size=1.0):
        self.pos=x0
        self.ndim=ndim
        self.possibleSteps=[]
        for i in range(ndim):
            step=numpy.zeros(ndim)
            step[i]= - step_size
            self.possibleSteps.append(numpy.array(step,dtype='f'))
            step[i]= + step_size
            self.possibleSteps.append(step.copy())
        self.npossible=len(self.possibleSteps)

    def pickStep(self):
        istep = numpy.random.choice(range(self.npossible))
        return self.possibleSteps[istep]
        
    def doSteps(self,n):
        positions=numpy.ndarray((n+1,self.ndim),dtype='f')
        for i in range (0, n+1):
            positions[i] = self.pos
            self.pos = positions[i] + self.pickStep()
        return positions
```

```python
# this test is worth 2 marks
numpy.random.seed(1111)
w = walker(numpy.zeros(1))
pos_test = w.doSteps(10)
reference = [[ 0.], [-1.], [ 0.], [ 1.], [ 2.], [ 1.], [ 0.], [-1.], [-2.], [-3.], [-4.]]
assert len(pos_test)==11
# plots to help debugging
plt.plot(range(11),pos_test, label='your trajectory')
plt.plot(range(11),reference,'o', label='reference points')
plt.legend()
plt.xlabel('step number')
assert (pos_test == reference).all()
```

**Task 1**

Make a plot of average position and average squared position of 100 1D walkers using 1000 steps. Your plot needs a legend, title and labels. [5 marks]


```python
numpy.random.seed(1114)

#### I have used the formula for the average squared distance from the mean from the lecture notes.
#### I presumed this was what Task 1 and Task 2 were asking as this was what covered in the lecture.
#### Even if we are not expected to minus the mean, as the mean is approximately 0 (the number of walkers is large), the result would be the same as displayed.

seed_using = 1114

nwalkers = 100
nsteps = 1000
ndim = 1



def mean_array(number_of_walkers, number_of_steps, dimensions, seedvalue):
    numpy.random.seed(seedvalue)
    summation = 0
    for i in range(number_of_walkers):
        w = walker(numpy.zeros(dimensions), ndim = dimensions)
        ys = w.doSteps(number_of_steps)
        summation = summation + ys
    return summation/number_of_walkers

def avg(number_of_walkers, number_of_steps, dimensions, seedvalue):
    mean_array_for_this = mean_array(number_of_walkers, number_of_steps, dimensions, seedvalue)
    numpy.random.seed(seedvalue)
    summation = 0
    for i in range(number_of_walkers):
        w = walker(numpy.zeros(dimensions), ndim = dimensions)
        individual_array = w.doSteps(number_of_steps)
        diff_array = individual_array - mean_array_for_this
        ESD_array = []
        for j in range(0,number_of_steps+1):
            square_distance = np.dot(diff_array[j],diff_array[j])
            ESD_array.append(square_distance)
        summation = np.array(ESD_array) + summation
    return summation/number_of_walkers


SD_values = avg(nwalkers, nsteps, ndim, seed_using)
mean_values = mean_array(nwalkers, nsteps, ndim, seed_using)
plt.plot(range(nsteps+1),SD_values, color = 'orange', label = 'Average Squared Position')
plt.plot(range(nsteps+1),mean_values, color = 'blue', label = 'Mean Position')
plt.legend(loc = 'upper left')
plt.xlabel('Step number')
plt.ylabel('Position variable')
plt.title('Mean position and average squared\nposition of 100 1D walkers')  ##Position variables of a random 1D walker\n(from sample of 100 walkers) 
plt.show()
```

## 2D walkers

Use 1000 walkers randomly distributed in the unit square (the positions are given in the array `rand_pos`) simulate the diffusion of particles with step size 0.05. Make a plot of the position of the walkers after 10, 100 and 500 steps. The plots should have labels and titles.

Tip: Use `plt.scatter` and consider using its `alpha` option to make you plot look nicer. [6 marks]


```python
dimension_using = 2
nwalkers=1000
numpy.random.seed(1114)

rand_pos = numpy.random.uniform(size=(nwalkers, dimension_using))

x_pos_10 = []
y_pos_10 = []
x_pos_100 = []
y_pos_100 = []
x_pos_500 = []
y_pos_500 = []

for j in range(nwalkers):
        w = walker(rand_pos[j], ndim = dimension_using, step_size = 0.05)
        big_array = w.doSteps(500)
        x_pos_10.append(big_array[10,0])
        y_pos_10.append(big_array[10,1])
        x_pos_100.append(big_array[100,0])
        y_pos_100.append(big_array[100,1])
        x_pos_500.append(big_array[500,0])
        y_pos_500.append(big_array[500,1])
        

matrix_x = np.vstack((np.array(x_pos_10), np.array(x_pos_100), np.array(x_pos_500)))
matrix_y = np.vstack((np.array(y_pos_10), np.array(y_pos_100), np.array(y_pos_500)))
    
# this code is a template to create 3 subplots in one, adapt it to your task!
plt.figure(figsize=(18,6))
for i, nsteps in enumerate([10,100,500]):
    plt.subplot(131+i)
    plt.xlim((-3, 4))
    plt.ylim((-3, 4))
    plt.xlabel('x position of walker')
    plt.ylabel('y position of walker')
    plt.title("Walker diffusion after {} steps ".format(nsteps))
    plt.scatter(matrix_x[i,:], matrix_y[i,:], alpha=0.1)
```
