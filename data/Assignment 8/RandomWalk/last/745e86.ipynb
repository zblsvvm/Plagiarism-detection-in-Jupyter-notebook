## Random Walks

This assignment will look at some properties of random walks.

```python
import numpy
from matplotlib import pyplot as plt 
```

To do our work we will implement a walker class. When initialised a list of possible steps is populated. In one dimension it is

[+s] , [-s] 

where s is the step size, it defaults to 1 but can be set as an argument in the constructor. In two dimensions the steps list contains

[ +s , 0 ] , [ -s , 0 ] ,  [ 0 , +s ] , [ 0 , -s ]

At each step the current position of the walker, saved in `self.pos`, is updated by adding one of the possible steps. The function `pickStep` chooses randomly one of the possible steps. Use this function to implement the `doSteps` function that performs `n` steps and returns a `(n+1) x ndim` array representing the trajectory of the walker, including the starting point. 

```python

class walker:
    def __init__(self,x0,ndim=1, step_size=1.0):
        self.pos=x0
        self.ndim=ndim
        self.possibleSteps=[]
        for i in range(ndim):
            step=numpy.zeros(ndim)
            step[i]= - step_size
            self.possibleSteps.append(numpy.array(step,dtype='f'))
            step[i]= + step_size
            self.possibleSteps.append(step.copy())
        self.npossible=len(self.possibleSteps)

    def pickStep(self):
        istep = numpy.random.choice(range(self.npossible))
        return self.possibleSteps[istep]
        
    def doSteps(self,n):
        positions=numpy.ndarray((n+1,self.ndim),dtype='f')
        
        positions[0,:] = numpy.array([0]*self.ndim)
        for i in range(n):
            positions[i+1,:] = positions[i,:] + numpy.array([self.pickStep()])
        
        return positions
    
```

```python
# this test is worth 2 marks
numpy.random.seed(1111)
w = walker(numpy.zeros(1))
pos_test = w.doSteps(10)
reference = [[ 0.], [-1.], [ 0.], [ 1.], [ 2.], [ 1.], [ 0.], [-1.], [-2.], [-3.], [-4.]]
assert len(pos_test)==11
# plots to help debugging
plt.plot(range(11),pos_test, label='your trajectory')
plt.plot(range(11),reference,'o', label='reference points')
plt.legend()
plt.xlabel('step number')
assert (pos_test == reference).all()
```

**Task 1**

Make a plot of average position and average squared position of 100 1D walkers using 1000 steps. Your plot needs a legend, title and labels. [5 marks]


```python
nsteps = 1000
ndim = 1
nwalkers = 100

total = numpy.ndarray((nsteps+1,nwalkers),dtype='f')
av = numpy.ndarray((nsteps+1),dtype='f')
avsquared = numpy.ndarray((nsteps+1),dtype='f')

for i in range(nwalkers):
    w = walker(numpy.zeros(1), ndim)
    total[:,i] = w.doSteps(nsteps)[:,0]

c=0
for i in range(nsteps+1):
    av[i] = numpy.mean(total[i,:])
    for j in range(nwalkers):
        c = c + total[i][j]**2
        
    avsquared[i] = c/nwalkers
    c = 0

plt.plot(range(nsteps+1), av, label='Average Position')
plt.plot(range(nsteps+1), avsquared, label='Average Squared Position')
plt.legend()   
plt.title('Graph of the average and average squared position of 100 1D walkers')
plt.xlabel('Steps')
plt.ylabel('Value')
    
```

**Task 2**

Make a plot to show that the average squared distance scaling is independent of the dimension in which the walker moves. Use 100 steps and 400 walkers and use $D=1,2,3,4$. The plot should have a title, legend and labels. [5 marks]


```python
nsteps = 100
nwalkers = 400
dimensions = range(1,5)

total1 = numpy.ndarray((nsteps+1,nwalkers, 1),dtype='f')
total2 = numpy.ndarray((nsteps+1,nwalkers, 2),dtype='f')
total3 = numpy.ndarray((nsteps+1,nwalkers, 3),dtype='f')
total4 = numpy.ndarray((nsteps+1,nwalkers, 4),dtype='f')

avsqudis = numpy.ndarray((nsteps+1, 4),dtype='f')

for i in range(nwalkers):
    w = walker(numpy.zeros(1), 1)
    total1[:,i,:] = w.doSteps(nsteps)
    x = walker(numpy.zeros(2), 2)
    total2[:,i,:] = x.doSteps(nsteps)
    y = walker(numpy.zeros(3), 3)
    total3[:,i,:] = y.doSteps(nsteps)
    z = walker(numpy.zeros(4), 4)
    total4[:,i,:] = z.doSteps(nsteps)

d=0
e=0
f=0
g=0
for i in range(nsteps+1):
    for j in range(nwalkers):
        d = d + sum(total1[i][j][:]**2)
        e = e + sum(total2[i][j][:]**2)
        f = f + sum(total3[i][j][:]**2)
        g = g + sum(total4[i][j][:]**2)
    avsqudis[i][0] = d/nwalkers
    avsqudis[i][1] = e/nwalkers
    avsqudis[i][2] = f/nwalkers
    avsqudis[i][3] = g/nwalkers
        
plt.plot(range(nsteps+1), avsqudis[:,0], label='1D')
plt.plot(range(nsteps+1), avsqudis[:,1], label='2D')
plt.plot(range(nsteps+1), avsqudis[:,2], label='3D')
plt.plot(range(nsteps+1), avsqudis[:,3], label='4D')

plt.legend()
plt.title('Average squared distance against steps taken for 400 walkers in different dimensions')
plt.xlabel('Steps')
plt.ylabel('Averaged Squared Distance')
```
