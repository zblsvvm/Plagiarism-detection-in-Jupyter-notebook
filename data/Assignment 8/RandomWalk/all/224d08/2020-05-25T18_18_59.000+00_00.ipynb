## Random Walks

This assignment will look at some properties of random walks.

```python
import numpy
from matplotlib import pyplot as plt 
```

To do our work we will implement a walker class. When initialised a list of possible steps is populated. In one dimension it is

[+s] , [-s] 

where s is the step size, it defaults to 1 but can be set as an argument in the constructor. In two dimensions the steps list contains

[ +s , 0 ] , [ -s , 0 ] ,  [ 0 , +s ] , [ 0 , -s ]

At each step the current position of the walker, saved in `self.pos`, is updated by adding one of the possible steps. The function `pickStep` chooses randomly one of the possible steps. Use this function to implement the `doSteps` function that performs `n` steps and returns a `(n+1) x ndim` array representing the trajectory of the walker, including the starting point. 

```python

class walker:
    def __init__(self,x0,ndim=1, step_size=1.0):
        self.pos=x0
        self.ndim=ndim
        self.possibleSteps=[]
        for i in range(ndim):
            step=numpy.zeros(ndim)
            step[i]= - step_size
            self.possibleSteps.append(numpy.array(step,dtype='f'))
            step[i]= + step_size
            self.possibleSteps.append(step.copy())
        self.npossible=len(self.possibleSteps)

    def pickStep(self):
        istep = numpy.random.choice(range(self.npossible))
        return self.possibleSteps[istep]
        
    def doSteps(self,n):
        positions=numpy.ndarray((n+1,self.ndim),dtype='f')
        positions[0] = self.pos
        for i in range(1, n+1):
            newpos = positions[i-1] + self.pickStep()
            positions[i] = newpos
            
        return positions
```

```python
# this test is worth 2 marks
numpy.random.seed(1111)
w = walker(numpy.zeros(1))
pos_test = w.doSteps(10)
reference = [[ 0.], [-1.], [ 0.], [ 1.], [ 2.], [ 1.], [ 0.], [-1.], [-2.], [-3.], [-4.]]
assert len(pos_test)==11
# plots to help debugging
plt.plot(range(11),pos_test, label='your trajectory')
plt.plot(range(11),reference,'o', label='reference points')
plt.legend()
plt.xlabel('step number')
assert (pos_test == reference).all()
```

**Task 1**

Make a plot of average position and average squared position of 100 1D walkers using 1000 steps. Your plot needs a legend, title and labels. [5 marks]


```python
nsteps = 1000
ndim = 1
nwalkers = 100
walker_paths = numpy.empty((nwalkers,nsteps+1))

for i in range(nwalkers):
    w = walker(numpy.zeros(1))
    ys = w.doSteps(nsteps)
    walker_paths[i] = numpy.transpose(ys)

averages = numpy.empty((2, nsteps+1))

for e in range(nsteps+1):
    averages[0, e] = numpy.average(walker_paths[:,e])
    averages[1, e] = numpy.average(walker_paths[:,e]**2)


plt.plot(range(nsteps+1), averages[0,:], label='Position')
plt.plot(range(nsteps+1), averages[1,:], label='Position Squared')
plt.xlabel('Steps')
plt.ylabel('Position')
plt.title('Plot of Average Position and \n Average Squared Position of 100 1D Walkers')
plt.legend()
```

**Task 2**

Make a plot to show that the average squared distance scaling is independent of the dimension in which the walker moves. Use 100 steps and 400 walkers and use $D=1,2,3,4$. The plot should have a title, legend and labels. [5 marks]


```python
nsteps = 100
nwalkers = 400
dimensions = range(1,5)

avdistsqrd = numpy.empty((len(dimensions), nsteps+1))

walker_paths1 = numpy.empty((nwalkers,nsteps+1, 1))
walker_paths2 = numpy.empty((nwalkers,nsteps+1, 2))
walker_paths3 = numpy.empty((nwalkers,nsteps+1, 3))
walker_paths4 = numpy.empty((nwalkers,nsteps+1, 4))

for e in range(nwalkers):
        w = walker(numpy.zeros(1), ndim=1)
        ys = w.doSteps(nsteps)
        walker_paths1[e] = ys
        
for e in range(nsteps+1):
    avdistsqrd[0, e] = numpy.average((numpy.linalg.norm(walker_paths1[:,e,:]))**2) 

for e in range(nwalkers):
        w = walker(numpy.zeros(1), ndim=2)
        ys = w.doSteps(nsteps)
        walker_paths2[e] = ys   
        
for e in range(nsteps+1):
    avdistsqrd[1, e] = numpy.average((numpy.linalg.norm(walker_paths2[:,e,:]))**2) 
    
    
for e in range(nwalkers):
        w = walker(numpy.zeros(1), ndim=3)
        ys = w.doSteps(nsteps)
        walker_paths3[e] = ys
        
for e in range(nsteps+1):
    avdistsqrd[2, e] = numpy.average((numpy.linalg.norm(walker_paths3[:,e,:]))**2) 

    
for e in range(nwalkers):
        w = walker(numpy.zeros(1), ndim=4)
        ys = w.doSteps(nsteps)
        walker_paths4[e] = ys

for e in range(nsteps+1):
    avdistsqrd[3, e] = numpy.average((numpy.linalg.norm(walker_paths4[:,e,:]))**2) 
    

for i in range(len(dimensions)):
    labels=['1D', '2D', '3D', '4D']
    plt.plot(range(nsteps+1), avdistsqrd[i,:], label=labels[i])
plt.legend()
plt.xlabel('Steps')
plt.ylabel('Distance')
plt.loglog()
plt.title('Logarithmic Plot of Average Squared Distance for Different Dimensions')
plt.show()
```

## 2D walkers

Use 1000 walkers randomly distributed in the unit square (the positions are given in the array `rand_pos`) simulate the diffusion of particles with step size 0.05. Make a plot of the position of the walkers after 10, 100 and 500 steps. The plots should have labels and titles.

Tip: Use `plt.scatter` and consider using its `alpha` option to make you plot look nicer. [6 marks]


```python
ndim=2
nwalkers=1000
nsteps=500

rand_pos = numpy.random.uniform(size=(nwalkers, ndim))
history = numpy.empty((nwalkers, nsteps+1, 2))

for e in range(nwalkers):
        w = walker(rand_pos[e], ndim, step_size=0.05)
        ys = w.doSteps(nsteps)
        history[e,:] = ys   

# this code is a template to create 3 subplots in one, adapt it to your task!
plt.figure(figsize=(18,6))
for i, nsteps in enumerate([10,100,500]): #counter, value
    plt.subplot(131+i)
    plt.xlim((-3, 4))
    plt.ylim((-3, 4))
    plt.title("Position of 1000 Randomly Distributed Particles \n After {} Steps".format(nsteps))    
    plt.scatter(history[:,nsteps,0], history[:,nsteps,1], alpha=0.10)
    plt.xlabel('x')
    plt.ylabel('y')
```
