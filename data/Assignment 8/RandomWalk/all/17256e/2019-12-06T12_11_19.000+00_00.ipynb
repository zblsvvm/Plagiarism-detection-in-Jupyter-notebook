## Random Walks

This assignment will look at some properties of random walks.

```python
import numpy
from matplotlib import pyplot as plt 
```

To do our work we will implement a walker class. When initialised a list of possible steps is populated. In one dimension it is

[+s] , [-s] 

where s is the step size, it defaults to 1 but can be set as an argument in the constructor. In two dimensions the steps list contains

[ +s , 0 ] , [ -s , 0 ] ,  [ 0 , +s ] , [ 0 , -s ]

At each step the current position of the walker, saved in `self.pos`, is updated by adding one of the possible steps. The function `pickStep` chooses randomly one of the possible steps. Use this function to implement the `doSteps` function that performs `n` steps and returns a `(n+1) x ndim` array representing the trajectory of the walker, including the starting point. 

```python

class walker:
    def __init__(self,x0,ndim=1, step_size=1):
        self.pos=x0
        self.ndim=ndim
        self.possibleSteps=[]
        for i in range(ndim):
            step=numpy.zeros(ndim)
            step[i]= - step_size
            self.possibleSteps.append(numpy.array(step,dtype='f'))
            step[i]= + step_size
            self.possibleSteps.append(step.copy())
        self.npossible=len(self.possibleSteps)

    def pickStep(self):
        istep = numpy.random.choice(range(self.npossible))
        return self.possibleSteps[istep]
        
    def doSteps(self,n):
        positions=numpy.ndarray((n+1,self.ndim),dtype='f')
        # YOUR CODE HERE
        
        for i in range(n+1):
            positions[i] = self.pos
            step = self.pickStep()
            self.pos += step
            
        return positions
```

```python
# this test is worth 2 marks
numpy.random.seed(1111)
w = walker(numpy.zeros(1))
pos_test = w.doSteps(10)
reference = [[ 0.], [-1.], [ 0.], [ 1.], [ 2.], [ 1.], [ 0.], [-1.], [-2.], [-3.], [-4.]]
assert len(pos_test)==11
# plots to help debugging
plt.plot(range(11),pos_test, label='your trajectory')
plt.plot(range(11),reference,'o', label='reference points')
plt.legend()
plt.xlabel('step number')
assert (pos_test == reference).all()
```

Make a plot of average position and average squared position of 100 1D walkers using 1000 steps. Your plot needs a legend, title and labels. [5 marks]


```python
nsteps = 1000
ndim = 1
nwalkers = 100

def avg_pos(steps, dims, paths):
    vals = numpy.zeros((steps+1)*paths)              # all values to take into account
    average_pos = []                                 # final values to be returned
    avg_pos_2 = []                                   # average position squared table
    for i in range(paths):                           # for number of walkers
        w = walker(numpy.zeros(1))
        pos_test = w.doSteps(steps)
        bah = len(pos_test)
        for g in range(bah):                         #adding each calculated value to vals
            vals[g + (i*bah)] = pos_test[g]
        
    
    for h in range(steps):                           #splicing arrays to get new arrays to average
        boop = vals[h::steps+1]
        boop2 = numpy.empty([len(boop)])
        cah = numpy.mean(boop)
        average_pos.append(cah)
        
        for k in range(len(boop)):                   #squaring each part of the position
            boop2[k] = (boop[k]**2)
        cah2 = numpy.mean(boop2)
        avg_pos_2.append(cah2)
        
    xs = []
    for x in range(steps):
        xs.append(x)
    
    return xs, average_pos, avg_pos_2

xs, ys, y2s = avg_pos(nsteps, ndim, nwalkers)        #avg position line


for i in range(nwalkers):                            #plotting the rest of the walked paths as comparison
    w = walker(numpy.zeros(1))
    yas = w.doSteps(nsteps)
    plt.plot(range(nsteps+1),yas)

plt.plot(xs, ys, 'k--', label = 'Average position')  # plotting avg position line
plt.plot(xs, y2s, 'b--', label = "Average squared position")
plt.legend(loc = 'upper left')                       # legend
plt.title("The average position and average position" 
          +" squared of random walked paths")
plt.xlabel("Number of steps")
plt.ylabel("Position")
```
