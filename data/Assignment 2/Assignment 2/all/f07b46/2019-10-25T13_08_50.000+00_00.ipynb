## Week 2: Numerical Integration

```python
import numpy 
import matplotlib.pyplot as plt
%matplotlib inline
```

Define the function `f`, such that $\textrm{f}(x) \equiv x^{2}\sin(x)$. This is the function that we will be integrating.

```python
def f(x):
    '''Function equivalent to x^2 sin(x).'''
    y1=(x**2)*numpy.sin(x)
    print(y1)
    return y1;
```

Ensure your function works with numpy arrays:

```python
xs=numpy.arange(0, 1, step=0.1)
assert numpy.isclose(f(xs), 
                     [0., 0.00099833, 0.00794677, 0.02659682, 0.06230693,
                      0.11985638, 0.20327129, 0.31566667, 0.4591079 , 0.6344948 ]).all()
```

Derive the indefinite integral of $f(x)$ nalytically. Call this function $g(x)$ and implement it below. Set the constant of integration such that $g(0)=0$.

```python
def g(x):
    '''Analytical integral of f(x).'''
    y2=-(x**2)*numpy.cos(x)+2*x*numpy.sin(x)+2*numpy.cos(x)-2
    print(y2)
    return y2;
```

Check your solution with the same numpy array:

```python
assert g(0) == 0.
```

```python
assert numpy.isclose(g(xs), 
                     [0., 0.00002497, 0.00039822, 0.00200482, 0.0062869, 
                      0.01519502, 0.03112138, 0.05681646, 0.09529087, 0.1497043 ]).all()
```

Now, using the analytically derived indefinite integral, $g(x)$, define a function which calculates the definite integral of $f(x)$ over the interval $(x_{min},~x_{max})$.

```python
def integrate_analytic(xmin, xmax):
    '''Analytical integral of f(x) from xmin to xmax.'''
    y3=g(xmax)-g(xmin)
    print(y3)
    return y3;

```

Check your analytic function:

```python
assert numpy.isclose(integrate_analytic(xmin=0, xmax=4), 1.096591)
```

## Numerical implementation

Create a function which calculates the definite integral of the function $f(x)$ over the interval $(x_{min},~x_{max})$ using Simpson's rule with $N$ panels.

```python
def integrate_numeric(xmin, xmax, N):
    ''' 
    Numerical integral of f from xmin to xmax using Simpson's rule with 
        N panels.
    '''
    
    Vec = numpy.zeros(N)
    Vec[0] = xmin
    dx = (xmax-xmin)/N

    for i in range (0,N): 

         Vec[i] = (dx/6)*((f(xmin+(i*dx))) + (4*f(xmin+(i*dx)+(dx/2))) + (f(xmin+(i*dx)+dx)))

    Integ = sum(Vec)
    print(Integ)
    return Integ;
                          


                        
                         
                    
 



```

Make sure you have implemented Simpson's rule correctly:

```python
assert numpy.isclose(integrate_numeric(xmin=0, xmax=4, N=1), 1.6266126)
```

```python
assert numpy.isclose(integrate_numeric(xmin=0, xmax=4, N=50), 1.096591)
```

There will always be some discrepancy between a numerically calculated result and an analytically derived result. Produce a log-log plot showing the fractional error between these two results as the number of panels is varied. The plot should have labels and a title.


```python
x0, x1 = 0, 2  # Bounds to integrate f(x) over
panel_counts = [4, 8, 16, 32, 64, 128, 256, 512, 1024]  # Panel numbers to use
result_analytic = integrate_analytic(x0, x1)# Define reference value from analytical solution
result_numerical = numpy.zeros(len(panel_counts))
Y1 = numpy.zeros((len(panel_counts),1))
for i in range (0, len(panel_counts)):
    Y1[i] = result_analytic - integrate_numeric(0, 2, panel_counts[i]) 
plt.figure(figsize=(8, 4))
plt.loglog(panel_counts, Y1)
plt.xlabel("Panel Count")
plt.ylabel("Error")
plt.title('Error Between Numerical and Analytical with Varying Panel Number')
```

What effect(s) does changing the number of panels used have
on the accuracy of the numerical method? What happens if the number of panels is taken too large?

Increasing panel number increases the accuracy of the Simpson integration. This is due to the smaller step length, and so less of the curve is ignored when calculating the integral. If the number of panels is too large the difference between different parts of the Simpsons function become too similar. This makes their difference too small and so the accuracy of the result is less, since the computer can only store so many decimal places.
